```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from Fred import Fred
```

```python
sns.color_palette("Paired")
```

```python
df = Fred.run(500)
```

```python
c_palette = sns.blend_palette(
    ['red', 'orange', 'green', 'blue'],
    n_colors=len(df.c.unique())
)

sns.set_style('whitegrid')

c_palette = { x: 'gray' for x in df.c.unique() }

red  = (0.5, 0.0, 0.0, 1.0)
cyan = (0.0, 0.35, 0.35, 1.0)

c_palette[0] = cyan
c_palette[1] = red
c_palette[2] = red
c_palette[3] = cyan

sns.kdeplot(
    x=df.x,
    y=df.y,
    hue=df.d,
    bw_adjust=1.5,
    levels=5,
    warn_singular=False,
    fill=False,
    legend=False,
    palette=c_palette,
    alpha=0.5
)

sns.scatterplot(
    x=df.x,
    y=df.y,
    hue=df.d,
    s=2,
    palette=c_palette,
    alpha=0.5
)
```
